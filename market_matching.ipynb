```python
import io
text = """2
10
buy 5 shares at 100
buy 5 shares at 105
sell 3 shares at 102
sell 5 shares at 105
buy 10 shares at 101
sell 12 shares at 100
buy 2 shares at 103
sell 2 shares at 104
buy 1 shares at 102
sell 1 shares at 103
12
buy 10 shares at 200
sell 5 shares at 210
buy 5 shares at 205
sell 8 shares at 205
buy 10 shares at 208
sell 2 shares at 207
buy 5 shares at 210
sell 5 shares at 210
sell 5 shares at 215
buy 3 shares at 212
sell 3 shares at 212
buy 2 shares at 213"""

f = io.StringIO(text) #字串變成io
T = int(f.readline()) # T=2 表示兩天的資料
for t in range(T):
  N = int(f.readline())
  buy = []
  sell = []
  stockprice = -1 #成交價
  for n in range(N): #每一筆委託檔讀取
    str = f.readline().split() # str = buy 10 shares at 100
    share = int(str[1]) # 10
    price = int(str[-1]) # 100
    if str[0] == "buy":
      while len(sell) > 0: #有賣單
        order = sell[0] #目前最好的賣單
        if order[0] > price:
          break
        dealamount = min(order[1], share) #剩下的成交量
        stockprice = order[0] #成交價
        order[1] -= dealamount #賣單剩餘量
        share -= dealamount
        if order[1] == 0:
          del sell[0]
        if share == 0:
          break
      if share>0: #還以剩，將直插入賣單
          i = 0
          while(i<len(buy) and price < buy[i][0]):
            i += 1
          if(i< len(buy) and price == buy[i][0]):
            buy[i][1] += share
          else:
            buy.insert(i, [price, share])
    else:
      while len(buy)>0:
        order = buy[0]
        if order[0] < price:
          break
        dealamount = min(order[1],share)
        # stockprice = order[0]
        stockprice = price #賣單成交價是自己價格
        order[1]-=dealamount
        share-=dealamount
        if order[1] == 0:
          del buy[0]
        if share == 0:
          break
      if share>0:
        i = 0
        while(i<len(sell) and price>sell[i][0]):
          i+=1
        if(i<len(sell) and price==sell[i][0]):
          sell[i][1] += share
        else:
          sell.insert(i,[price,share])


    # 印出
    if not sell:
      print('-',end=' ')
    else:
      print(sell[0][0], end=' ')
    if not buy:
      print('-',end=' ')
    else:
      print(buy[0][0], end=' ')
    if stockprice == -1:
      print('-')
    else:
      print(stockprice)

```
