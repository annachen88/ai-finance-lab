```python
import tensorflow as tf
from tensorflow.keras import layers, models, datasets
import numpy as np
(x_train,_),(x_test,_) = datasets.mnist.load_data()
x_train = x_train.astype('float32')/255.0
x_test = x_test.astype('float32')/255.0
latent_dim = 16
encoder = models.Sequential([
    layers.Input(shape=(28,28,1)),
    layers.Conv2D(16,3,strides=2,activation='relu', padding='same'),
    layers.Conv2D(32,3,strides=2,activation='relu', padding='same'),
    layers.Flatten(),
    layers.Dense(latent_dim)
])
decoder = models.Sequential([
    layers.Input(shape=(latent_dim,)),
    layers.Dense(7*7*32,activation='relu'),
    layers.Reshape((7,7,32)),
    layers.Conv2DTranspose(32,3,strides=2,activation='relu',padding='same'),
    layers.Conv2DTranspose(16,3,strides=2,activation='relu',padding='same'),
    layers.Conv2D(1,3,activation='sigmoid',padding='same')
])
autoencoder = models.Sequential([
    encoder,
    decoder
])
autoencoder.compile(optimizer='adam',loss='mse')
autoencoder.fit(x_train,x_train,epochs=10,batch_size=128,validation_data=(x_test,x_test))
```

```python
import imageio # animate
def create_interpolation_gif(imgs,steps,filename='transition.gif'):
  period = imgs.shape[0]-1
  vec = encoder.predict(imgs)
  images = []
  for i in range(period):
    # 每張圖片交錯
    vec1 = vec[i,:]
    vec2 = vec[i+1,:]
    interpolated_vecs = np.linspace(vec1,vec2,num=steps)
    interpolated_vecs = interpolated_vecs.reshape(steps,latent_dim)
    decoded_img = decoder.predict(interpolated_vecs)
    for j in range(steps):
      image = (decoded_img[j].reshape(28,28)*255).astype(np.uint8)
      images.append(image)
  imageio.mimsave(filename,images,duration=0.2,loop=0)

create_interpolation_gif(x_test[1000:1010],20)
```
